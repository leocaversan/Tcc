```python
import pandas as pd
import numpy as np

y_train = pd.read_csv('NSL-KDD/KDDTrain+.txt',header=None)
x_train = pd.read_csv('NSL-KDD/KDDTrain+.txt',header=None)
y_train
```

```python
x_train = x_train.iloc[:,4:-2]
x = x_train.values
x = x.astype(int)
x_train
```

```python
y_train = y_train.iloc[:,4:-1]
y_train[41] = np.where(y_train[41] == "normal", 0, 1)
y_train = y_train.iloc[:,-1:]
y_train = y_train.values
y = y_train.astype(int)
y_train
```

```python
from sklearn import metrics
from sklearn.discriminant_analysis import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
# from sklearn import metrics


x_train, x_test,  y_train, y_test = train_test_split(x, y,train_size=0.3)

clf = make_pipeline(StandardScaler(), SVC(gamma='auto'))
classification = clf.fit(x_train, y_train)
y_pred = classification.predict(x_test)
acuracy = metrics.accuracy_score(y_test, classification.predict(x_test))
predict = classification.score(x_test, y_test)
print(predict, acuracy)

# clf = tree.DecisionTreeRegressor()
# clf = clf.fit(x_test, y_train)
# tree.plot_tree(clf)

# clf = tree.DecisionTreeClassifier()
# clf = clf.fit(x_train, y_train)
# gnb.fit(x_train, y_train)
# previsoes = gnb.predict(x_test)
# acertos = metrics.accuracy_score(y_test,y_pred)
# performace = round(acertos,4)

```
