```python
import pandas as pd
import numpy as np

data_base = pd.read_csv('NSL-KDD/KDDTrain+.txt',header=None)
```

```python
x = data_base.iloc[:,4:-1]
x = x.iloc[:,:-1]
# x_train[41] = np.where(x_train[41] == "normal", 0, 1)
x = x.values
x = x.astype(int)



```

```python
y = data_base.iloc[:,4:-1]
y[41] = np.where(y[41] == "normal", 0, 1)
y = y.iloc[:,-1:]
y = y.values
y = y.astype(int)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics

x_train, x_test,  y_train, y_test = train_test_split(x, y,train_size=0.3)

knn = KNeighborsClassifier (n_neighbors=5)
knn.fit(x_train, y_train)
y_pred = knn.predict(x_test)
acuracy = metrics.accuracy_score(y_test,y_pred)
performace_values = round(acuracy,4)


print(performace_values)
```

```python
from sklearn.svm import SVC

import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

classification = SVC(random_state=0)
classification.fit(x_test, y_test)

 
color = 'white'
cm = confusion_matrix(y_test,y_pred)

disp = ConfusionMatrixDisplay(confusion_matrix=cm, 
                              display_labels=classification.classes_)


disp.plot()
plt.show()
```
